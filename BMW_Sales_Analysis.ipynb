```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# 1. Prepare total yearly sales data
df_total = df.groupby('Year')['Sales_Volume'].sum().reset_index()
df_total['ds'] = pd.to_datetime(df_total['Year'], format='%Y')
df_total = df_total.rename(columns={'Sales_Volume': 'y'})

# 2. Create and fit the Prophet model
model = Prophet(yearly_seasonality=True)
model.fit(df_total[['ds', 'y']])

# 3. Create future dataframe till 2035
last_year = df_total['ds'].dt.year.max()
future = model.make_future_dataframe(periods=(2035 - last_year), freq='Y')

# 4. Forecast
forecast = model.predict(future)

# 5. Plot forecast as line chart
plt.figure(figsize=(10, 6))
plt.plot(df_total['ds'], df_total['y'], label='Historical Sales', linewidth=2)
plt.plot(forecast['ds'], forecast['yhat'], label='Forecasted Sales', linestyle='--', color='orange', linewidth=2)
plt.fill_between(forecast['ds'], forecast['yhat_lower'], forecast['yhat_upper'], color='orange', alpha=0.2)

plt.title('BMW Total Sales Forecast till 2035')
plt.xlabel('Year')
plt.ylabel('Sales Volume')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Prepare data
df_model = df[['Model', 'Price_USD', 'Sales_Volume']].dropna()

# Store results
scenario_results = []

models = df_model['Model'].unique()

for model in models:
    df_sub = df_model[df_model['Model'] == model]

    if len(df_sub) < 5:
        continue  # skip if not enough data

    # Train regression model
    X = df_sub[['Price_USD']]
    y = df_sub['Sales_Volume']
    reg = LinearRegression()
    reg.fit(X, y)

    # Get current average price
    current_price = df_sub['Price_USD'].mean()

    # Scenarios
    scenarios = {
        'Optimistic (-10%)': current_price * 0.9,
        'Base Case': current_price,
        'Pessimistic (+10%)': current_price * 1.1
    }

    for scenario, price in scenarios.items():
        predicted_sales = reg.predict([[price]])[0]
        scenario_results.append({
            'Model': model,
            'Scenario': scenario,
            'Price': price,
            'Predicted_Sales': predicted_sales
        })

# Create DataFrame
df_scenarios = pd.DataFrame(scenario_results)

# Visualize
import seaborn as sns
plt.figure(figsize=(14, 6))
sns.barplot(data=df_scenarios, x='Model', y='Predicted_Sales', hue='Scenario')
plt.title('Sales Predictions under Different Price Scenarios for Each Model')
plt.xticks(rotation=45)
plt.ylabel("Predicted Sales Volume")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Prepare data
df_model = df[['Model', 'Price_USD', 'Sales_Volume']].dropna()

# Store results
scenario_results = []

models = df_model['Model'].unique()

for model in models:
    df_sub = df_model[df_model['Model'] == model]

    if len(df_sub) < 5:
        continue  # skip if not enough data

    # Train regression model
    X = df_sub[['Price_USD']]
    y = df_sub['Sales_Volume']
    reg = LinearRegression()
    reg.fit(X, y)

    # Get current average price
    current_price = df_sub['Price_USD'].mean()

    # Scenarios
    scenarios = {
    'Optimistic (-20%)': current_price * 0.8,
    'Base Case (No Change)': current_price,
    'Pessimistic (+20%)': current_price * 1.2
}

    for scenario, price in scenarios.items():
        predicted_sales = reg.predict([[price]])[0]
        scenario_results.append({
            'Model': model,
            'Scenario': scenario,
            'Price': price,
            'Predicted_Sales': predicted_sales
        })

# Create DataFrame
df_scenarios = pd.DataFrame(scenario_results)

# Visualize
import seaborn as sns
plt.figure(figsize=(14, 6))
sns.barplot(data=df_scenarios, x='Model', y='Predicted_Sales', hue='Scenario')
plt.title('Sales Predictions under Different Price Scenarios for Each Model')
plt.xticks(rotation=45)
plt.ylabel("Predicted Sales Volume")
plt.tight_layout()
plt.show()
```
